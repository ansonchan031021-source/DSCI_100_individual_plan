```r
options(repr.plot.width = 15, repr.plot.height = 7)
subscribe_count1 <- players |>
    filter(Age>12 & Age < 33)|>
    ggplot(aes(x = subscribe, fill = subscribe)) +
    geom_bar()+
    facet_grid(cols = vars(Age))+
    labs(title = "Subscription Status by Age (13~32 years old)",
        x = 'Subscription Status', y = 'Number of Players')+
    theme(text = element_text(size = 10))
subscribe_count1
```

In this data set, players under 14 and over 32 each have only one record. Therefore, before visualization, these cases as well as records with missing age values were removed to enhance clarity. Overall, the results indicate that subscribers are mainly concentrated in younger age groups, with 17-year-olds having the highest number of subscriptions. As age increases, the total number of players and subscribers gradually decreases. This suggests that younger players are more actively engaged in following news and community content related to the game.

```r
options(repr.plot.width = 15, repr.plot.height = 7)
subscribe_count2 <- players |>
    ggplot(aes(x = subscribe, fill = subscribe)) +
    geom_bar()+
    facet_grid(cols = vars(experience))+
    labs(title = "Subscription Status by Experience Level",
        x = 'Subscription Status', y = 'Number of Players')+
    theme(text = element_text(size = 20))
subscribe_count2
```

```r
options(repr.plot.width = 10, repr.plot.height = 10)
experience_played_hours <- players |>
    filter(played_hours < 10) |>
    ggplot(aes(x = experience, y = played_hours)) +
    geom_point(alpha = 0.3)+
    labs(title = "Played Hours by Experience (pleyed hours <10)",
        x = 'Experience Level', y = 'Played Hours')+
    theme(text = element_text(size = 25))

experience_played_hours
```

In this data set, since most players have total play hours under 10 hours, data beyond this threshold were excluded to focus on the primary distribution. As shown in the plot, regardless of experience level, most players spent less than 2 hours playing, with only a few reaching 6 to 7 hours. The overall distribution appears similar across all experience groups, showing no clear linear or consistent trend between experience level and play time.

# Methods and Plan

My question is: Can player `experience`, `played_hours`, and `age` predict whether a player subscribes to a game-related newsletter?  

This is a `knn classification problem`; we need to ensure all predictors are numerical before predicting. So we need to change words to numbers like: 1, 2, 3...  

We have 2 ways to change `experience` to numerical:  
+ **One-hot encoding**: This approach converts experience into multiple binary columns (e.g., Pro, Veteran, Amateur, Regular, Beginner), each taking values of 0 or 1.  
Limitation: It increases dimensionality, which can reduce model efficiency, especially with small datasets. The ranking from highest to lowest is Pro, Veteran, Amateur, Regular, and Beginner. However, this method can be expressed only as 0 and 1, and it cannot distinguish between them.
+ **Ordinal encoding**: This keeps experience as one numeric column, assigning ordered values such as Pro = 5, Veteran = 4, Amateur = 3, Regular = 2, Beginner = 1.  
Limitation: It introduces an artificial sense of order or distance between categories, which may not represent the real relationship among experience levels. The differences between each level may vary.

I will split the data into training (75%) and testing (25%) sets.

I will use Cross-validation and tuning k on the training set to determine the optimal value of k that minimizes classification error. After improving, we will evaluate on the testing set to assess predictive performance.
