# Part 1: Data Description  

#### There are 196 observations and 9 variables in the player table, the variables are:  

+ `experience`: charactor, it shows that the variable experience has five unique categories — `Pro`, `Veteran`, `Amateur`, `Regular`, and `Beginner` — representing different levels of player experience.
+ `subscribe`: logical, it shows whether the player subscribed to the game or not in `TRUE` or `FALSE`
+ `hashedEmail`: character, it stores anonymized user email identifiers in a hashed format to protect personal information.
+ `played_hours	`: double, it records the game playtime in hours, with a range from 0 to 223.1.
+ `name`: character, it shows different players' names.
+ `gender`: charactor, it shows the variable genders have 7 unique categories - `Male`, `Female`, `Non-binary`, `Prefer not to say`, `Agender`, `Two-Spirited` and `other`. 
+ `Age`: double, it records the players' age, with a range of 9 to 58.

#### There are 1535 observations and 5 variables in the sessions table, the variables are:  

+ `hashedEmail`: character, it stores anonymized user email identifiers in a hashed format to protect personal information.
+ `start_time`: character, it shows the game start time, each entry follows the date-time format DD/MM/YYYY (24-hour clock).
+ `end_time`: character, it shows the game end time, each entry follows the date-time format DD/MM/YYYY (24-hour clock).
+ `original_start_time`: double, it is the original data and is not human-readable, so it cannot be used directly.
+ `original_end_time`: double, it is also the original data and is not human-readable, so it cannot be used directly.
